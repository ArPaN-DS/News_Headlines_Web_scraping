```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager

# Set Chrome options
options = Options()
options.add_argument("--headless")  # Run Chrome in headless mode
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")
# Ensure the binary location is correct
options.binary_location = "/usr/bin/google-chrome"

# Initialize the WebDriver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
driver.get("https://timesofindia.indiatimes.com/topic/supreme-court-of-india")

# Let the page load completely
driver.implicitly_wait(5)

# Extract headlines using CSS selector for multiple classes
headlines = driver.find_elements(By.CSS_SELECTOR, "fHv_i o58kM")

# Print the headlines
for idx, headline in enumerate(headlines, start=1):
    print(f"{idx}. {headline.text.strip()}")

# Close the driver
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from webdriver_manager.chrome import ChromeDriverManager
import time

# Set up options for headless mode
options = Options()
options.add_argument('--headless')  # Run headless Chrome (no GUI)
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')

# Initialize the WebDriver using webdriver-manager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Open the Times of India page
driver.get("https://timesofindia.indiatimes.com/topic/supreme-court-of-india")

# Allow the page to load
time.sleep(5)

# Find and click the "Load More Articles" button multiple times if present
load_more_button = None
while True:
    try:
        load_more_button = driver.find_element(By.XPATH, '//button[contains(text(), "Load More Articles")]')
        load_more_button.click()
        time.sleep(3)  # Wait for the page to load more articles
    except:
        # If no "Load More" button is found, exit the loop
        break

# Extract the headlines
headlines = driver.find_elements(By.CLASS_NAME, "fHv_i")

# Print the headlines
for idx, headline in enumerate(headlines, start=1):
    print(f"{idx}. {headline.text.strip()}")

# Close the driver
driver.quit()
```

```python
import pandas as pd  # For saving data to Excel

# Store the headlines in a list (you already have this part)
headlines_list = [headline.text.strip() for headline in headlines]

# Create a DataFrame to store the headlines
df = pd.DataFrame(headlines_list, columns=["Headline"])

# Save the DataFrame to an Excel file
df.to_excel("supreme_court_headlines.xlsx", index=False)

print("Headlines saved to 'supreme_court_headlines.xlsx'.")
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the page you want to scrape
url = 'https://timesofindia.indiatimes.com/topic/supreme-court-of-india'

# Send a GET request to fetch the page content
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find all headlines in the specified div class
    headlines = soup.find_all('div', class_='fHv_i o58kM')

    # Extract the text from each headline and print it
    if headlines:
        for index, headline in enumerate(headlines, start=1):
            print(f"{index}. {headline.get_text(strip=True)}")
    else:
        print("No headlines found.")
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
import time
from bs4 import BeautifulSoup

# Set up headless Chrome options
options = Options()
options.headless = True

# Initialize WebDriver with the options
driver = webdriver.Chrome(options=options)

# Open the URL
url = 'https://timesofindia.indiatimes.com/topic/supreme-court-of-india'
driver.get(url)

# Wait for the page to load
time.sleep(3)

# Simulate clicking the "Load More" button
while True:
    try:
        load_more_button = driver.find_element(By.XPATH, "//button[contains(text(), 'Load More Articles')]")
        ActionChains(driver).move_to_element(load_more_button).click().perform()
        time.sleep(2)  # Wait for new content to load
    except:
        break  # Break the loop if the "Load More" button is no longer found

# Once all content is loaded, scrape the headlines
soup = BeautifulSoup(driver.page_source, 'html.parser')
headlines = soup.find_all('div', class_='fHv_i o58kM ')

# Print the headlines
for index, headline in enumerate(headlines, start=1):
    print(f"{index}. {headline.get_text(strip=True)}")

# Close the driver
driver.quit()
```
